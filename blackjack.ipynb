```python
suits = ("Spade","Club","Diamond","Heart")
numbers  = ("Two","Three","Four","Five","Six","Seven","Eight","Nine","Ten","Jack","Queen","King","Ace")
values = {"Two":2,"Three":3,"Four":4,"Five":5,"Six":6,"Seven":7,"Eight":8,"Nine":9,"Ten":10,"Jack":10,"Queen":10,"King":10,"Ace":11}

import random
```

```python
class Cards:
    def __init__(self,num,su):
        self.num = num
        self.su = su
    
    
    def __str__(self):
        return f'{self.num} of {self.su}'
```

```python
class Deck:
    def __init__(self):
        self.deck=[]
        for su in suits:
            for num in numbers:
                self.deck.append(Cards(num,su))
                
    def __str__(self):
        main_deck = ''
        for c in self.deck:
            main_deck += '\n' + c.__str__()
        return "The deck is:" + main_deck  
    
    def shuffle(self):
        random.shuffle(self.deck)
        
                
    def deal(self):
        single_card = self.deck.pop()
        return single_card
```

```python
class Hand:
    def __init__(self):
        self.hand =[]
        self.value = 0
        self.ace = 0

        
    def __str__(self):
        print (self.hand)
    
    def add_Card(self, card):
        self.hand.append(card)
        self.value += values[card.num]
        if card.num == "Ace":
            self.ace += 1
        return card
    
    def ace_adjust(self):
        while self.value > 21 and self.ace:
            self.value -= 10
            self.ace -= 1
```
