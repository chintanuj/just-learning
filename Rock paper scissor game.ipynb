```python
import random
import getpass
```

```python
lst = {"r":"Rock","s":"Scissor","p":"Paper"}
```

```python
def start_game():
    play1=input("Player 1, Enter your name: ")
    play2=input("Player 2, Enter your name: ")
    player1 = 0
    player2 = 0
    roundcount = 1
    maxscore = int(input("Please enter max points: "))
                     
    while player1!=maxscore and player2!=maxscore:
        print ("\n\n***Round ",roundcount," begins***\n")
        
        while True:
            d1 = getpass.getpass(f'{play1} >> Enter r for Rock, p for Paper, s for Scissor : ')
            if d1 in lst:
                while True:
                    d2 = getpass.getpass(f'{play2} >> Enter r for Rock, p for Paper, s for Scissor : ')
                    if d2 in lst:
                        break
                    else:
                        print (f'Invalid input! Please try again')
                break
            else:
                print (f'Invalid input! Please try again')
            
                        
        p1= lst [d1]
        p2= lst [d2]
        
        
        if p1 == "Rock" and p2=="Scissor":
            player1 += 1
            print (f'\n{play1} hand = {p1} \n{play2} hand = {p2} \n\n>>>>{play1} wins!<<<<')
            roundcount +=1
            
        elif p1 == "Scissor" and p2=="Rock":
            player2 += 1
            print (f'\n{play1} hand = {p1} \n{play2} hand = {p2} \n\n>>>>{play2} wins!<<<<')
            roundcount +=1
            
        elif p1 == "Paper" and p2=="Rock":
            player1 += 1
            print (f'\n{play1} hand = {p1} \n{play2} hand = {p2} \n\n>>>>{play1} wins!<<<<')
            roundcount +=1
            
            
        elif p1 == "Rock" and p2=="Paper":
            player2 += 1
            print (f'\n{play1} hand = {p1} \n{play2} hand = {p2} \n\n>>>>{play2} wins!<<<<')
            roundcount +=1
            
        elif p1 == "Scissor" and p2=="Paper":
            player1 += 1
            print (f'\n{play1} hand = {p1} \n{play2} hand = {p2} \n\n>>>>{play1} wins!<<<<')
            roundcount +=1
            
            
        elif p1 == "Paper" and p2=="Scissor":
            player2 += 1  
            print (f'\n{play1} hand = {p1} \n{play2} hand = {p2} \n\n>>>>{play2} wins!<<<<')
            roundcount +=1
                        
        elif p1 == p2:
            print (f'\n{play1} hand = {p1} \n{play2} hand = {p2} \n\n>>>>Its a Draw!<<<<')
            roundcount +=1
            
        else:
            print (f'Invalid input. Please enter r for Rock, p for Paper, s for Scissor')
            roundcount +=1
        
        
        
        
    if player1 == maxscore:
        print ("____________________________\n\n",f'Game Over! {play1} Wins!\n\n{play1} score is {player1}\n{play2} score is {player2}')
    
    else:
        print ("____________________________\n\n",f'Game Over! {play2} Wins!\n\n{play1} score is {player1}\n{play2} score is {player2}')
```

```python
start_game()
```
